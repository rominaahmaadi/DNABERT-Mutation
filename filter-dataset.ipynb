```python
import pandas as pd
```

```python
# 1. Specify columns to load
cols = [
    "Name", "ClinicalSignificance", "Assembly", "Chromosome",
    "Start", "Stop", "ReferenceAllele", "AlternateAllele",
    "GeneSymbol", "ReviewStatus"
]

# 2. Assign memory-efficient 'category' types to repetitive columns
dtypes_map = {
    "ClinicalSignificance": "category",
    "Assembly": "category",
    "Chromosome": "category",
    "ReviewStatus": "category"
}

# 3. Read specified columns
df = pd.read_csv(
    "variant_summary.txt.gz",
    sep="\t",
    usecols=cols,
    dtype=dtypes_map
)

print(f"Initial rows: {df.shape[0]}")

# 4. Filter for Assembly, Significance, and Review Status
df = df[
    (df["Assembly"] == "GRCh38") &
    (df["ClinicalSignificance"].isin(["Pathogenic", "Benign"])) &
    (df["ReviewStatus"].isin([
        "reviewed by expert panel",
        "criteria provided, multiple submitters, no conflicts"
    ]))
]

# Removed the filter that was excluding 'na' values since they are prevalent in this set
print(f"Final filtered rows: {df.shape[0]}")
```

```python
df.to_csv("filtered_variant_summary.csv", index=False)
```
